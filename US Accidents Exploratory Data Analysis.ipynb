Currently, there are about 2.8 million accident records in this dataset.

### Start Time


```python
# Changing data type to date time format
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
# To check the format
df.Start_Time[0]
```

```python
# Find hour of the day 
df.Start_Time.dt.hour
```

```python
sns.histplot(df.Start_Time.dt.hour, bins=24)
# There is 
```

```python
# Percentage of accidents (Hour) for better understanding of Data 
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

* A high percentage of accidents occur between 6 am to 10 am (probably people in a hurry to get to work)
* Next higest percentage is 3 pm to 6 pm. This is obviously peak hours when people are traveling back to home.  
* Percenatge of accidents decreases during off-peak hours from 7 pm to 10 pm. 

```python
#Percentage of accidents (Day of the week)
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

Percentage of accidents tend to be less during weekends, because people are not traveling to work.

Is the distribution of accidents by hour the same on weekends as on weekdays?

```python
#Let's check weekend (SUNDAY)
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

On sunday the graph shows more evenly distribution. On Sunday during afternoon more accident occur, as people are traveling places.

```python
#Let's check weekday (SUNDAY)
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

Comparing both graphs, On Sundays, the peak occurs between 10 am and 3 pm, unlike weekdays, for instance 6am to 9 am, people are going to work, increase in percentage of accidents and again the percentage of accidents increased when people are going back home from work, 3pm to 5pm.

```python
# Percentage of accidents (month)
sns.distplot(df.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

During summer the Percentage of accidents tend to be low compare to the end of the year. This could be during winter roads are more challenging to drive due to poor weather conditions. Therefore, more Percentage of accidents occur during Oct to Dec. 

```python
#Looking at data from 2019
df_2019 = df[df.Start_Time.dt.year == 2019]
sns.distplot(df_2019.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```
